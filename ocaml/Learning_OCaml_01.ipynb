## Chapter 1, basics are first!

In this chapter we learnt the basics about programming in general. For example, OCaml is _an expression language_, **values** are **expressions** but they cannot be reduced further.

```ocaml
1 + 2 ;;
```

This will return

```ocaml
- : int = 3
```

And it means the _expression `1+2`_ returns the _value `3`_ of type **int**.

The operators `mod, /, *` have precedence over the operators `+, -` and the binary operator `&&` has precedence over `||` but both boolean operators has no precedence over `+, -, mod, /, *`
